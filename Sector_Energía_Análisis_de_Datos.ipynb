```python
#Obtener binarios por columna de tiempo y valores
dummies_datos = pd.get_dummies(df_getopeale['estado'])
#dummies_datos.head()
df1 = pd.concat((df_getopeale,dummies_datos), axis=1)
df1.head()
```

```python
#Tiempos activos por cada estado
df1['deltat_green[segundos]'] = df1['deltat'] * df1['green']
df1['deltat_offline[segundos]'] = df1['deltat'] * df1['offline']
df1['deltat_orange[segundos]'] = df1['deltat'] * df1['orange']
df1['deltat_yellow[segundos]'] = df1['deltat'] * df1['yellow']
#Valores de potencia por cada estado
df1['valor_green[kwh]'] = df1['valor'] * df1['green']
df1['valor_offline[kwh]'] = df1['valor'] * df1['offline']
df1['valor_orange[kwh]'] = df1['valor'] * df1['orange']
df1['valor_yellow[kwh]'] = df1['valor'] * df1['yellow']
#Consumos

df1['consumo_green[kwh]']= df1['valor_green[kwh]']*(df1['deltat_green[segundos]']/3600)
df1['consumo_offline[kwh]'] = df1['valor_offline[kwh]']*(df1['deltat_offline[segundos]']/3600)
df1['consumo_orange[kwh]'] = df1['valor_orange[kwh]']*(df1['deltat_orange[segundos]']/3600)
df1['consumo_yellow[kwh]'] = df1['valor_yellow[kwh]']*(df1['deltat_yellow[segundos]']/3600)
#Totales
df1['total_tiempo'] = df1['deltat_green[segundos]'] + df1['deltat_offline[segundos]'] + df1['deltat_orange[segundos]'] + df1['deltat_yellow[segundos]']
df1['total_consumo[kwh]'] = df1['consumo_green[kwh]'] + df1['consumo_offline[kwh]'] + df1['consumo_orange[kwh]'] + df1['consumo_yellow[kwh]']
```

```python
#Eliminar columnas
df1 = df1.drop(["valor",	"estado",	"valor_raw", "deltat",	"green",	"offline","orange", "yellow"], axis =1)
df1.head()
```

```python
import json

def main():
    myfile = "/content/drive/MyDrive/Data science/Desafío Eclariti/alepa_getconsumo.json"
    with open(myfile,"r") as f:
        d = json.load(f)
        for i in d:
            if "consumo" in d[i]["KWH_ACT"] and len(d[i]["KWH_ACT"]["consumo"]["actual"]) > 0:
                df = pd.Series(d[i]["KWH_ACT"]["consumo"]["actual"]["fechahora"]).to_frame(name="KWH_ACT").reset_index().rename(columns={"index":"fechahora"})
                print(df)
                # Save DataFrame to CSV
                df.to_csv("/content/drive/MyDrive/Data science/Desafío Eclariti/alepa_getconsumo.csv", index=False)


if __name__ == "__main__":
    main()
```
